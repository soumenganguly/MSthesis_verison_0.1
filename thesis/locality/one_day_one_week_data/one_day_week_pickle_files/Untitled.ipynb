```python
import pickle

pul_pdf_one_week = open('PUL_PDF_oneweek.pkl','rb')
p = pickle.load(pul_pdf_one_week)
```

```python
x = []
y = []
for words in p:
    x.append(p[words][0])
    y.append(p[words][1])
```

```python
import numpy as np
import matplotlib.pyplot as plt


# Choose how many bins you want here
num_bins = 10

xi=np.array(x)

# Use the histogram function to bin the data
counts, bin_edges = np.histogram(xi, bins=num_bins, normed=True)

# Now find the cdf
cdf = np.cumsum(counts)

# And finally plot the cdf
plt.plot(bin_edges[1:], cdf)

plt.show()
```
